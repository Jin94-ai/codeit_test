### 0. 경로/헬퍼 정의

````python
import os
import json
from pathlib import Path
from tqdm.notebook import tqdm
import pandas as pd
import shutil

# -----------------------------
# 경로 설정
# -----------------------------
ADDED_DATA_ROOT = "../data/added/ts4"
ADDED_TRAIN_IMG_DIR = os.path.join(ADDED_DATA_ROOT, "train_images")       # TS_4_조합 (이미지)
ADDED_TRAIN_ANNO_DIR = os.path.join(ADDED_DATA_ROOT, "train_annotations") # TS_4_조합 라벨(JSON)

# TS4 JSON을 dl_idx/dl_name 규칙으로 고친 JSON 저장 폴더
ADDED_TRAIN_ANNO_FIXED_DIR = os.path.join(ADDED_DATA_ROOT, "train_annotations_fixed")

# 기존 raw 학습 데이터
RAW_DATA_ROOT = "../data"
RAW_TRAIN_IMG_DIR = os.path.join(RAW_DATA_ROOT, "train_images")
RAW_TRAIN_ANNO_DIR = os.path.join(RAW_DATA_ROOT, "train_annotations")

# 최종 통합 데이터 저장 경로
FIXED_DATA_ROOT = os.path.join(ADDED_DATA_ROOT, "fixed")
FIXED_TRAIN_IMG_DIR = os.path.join(FIXED_DATA_ROOT, "train_images")
FIXED_TRAIN_ANNO_DIR = os.path.join(FIXED_DATA_ROOT, "train_annotations")

os.makedirs(ADDED_TRAIN_ANNO_FIXED_DIR, exist_ok=True)
os.makedirs(FIXED_TRAIN_IMG_DIR, exist_ok=True)
os.makedirs(FIXED_TRAIN_ANNO_DIR, exist_ok=True)

print("ADDED_TRAIN_IMG_DIR:", ADDED_TRAIN_IMG_DIR)
print("ADDED_TRAIN_ANNO_DIR:", ADDED_TRAIN_ANNO_DIR)
print("ADDED_TRAIN_ANNO_FIXED_DIR:", ADDED_TRAIN_ANNO_FIXED_DIR)
print("RAW_TRAIN_ANNO_DIR:", RAW_TRAIN_ANNO_DIR)
print("FIXED_TRAIN_IMG_DIR:", FIXED_TRAIN_IMG_DIR)
print("FIXED_TRAIN_ANNO_DIR:", FIXED_TRAIN_ANNO_DIR)

# -----------------------------
# 헬퍼 함수
# -----------------------------
def find_json_files_recursively(root_dir):
    json_files = []
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith(".json"):
                json_files.append(os.path.join(dirpath, filename))
    return json_files
````

### 1. TS4 JSON별로 “실제 이미지 존재 여부” 먼저 검사

- JSON 안의 images[*].file_name 과 source_json_file 로부터 조합 폴더를 추출하고

- train_images/<조합>/<file_name> 이 실제 존재하는지 확인

- 이미지가 1장 이상 존재하는 JSON만 valid_json_paths 에 포함

````python
ts4_json_paths = find_json_files_recursively(ADDED_TRAIN_ANNO_DIR)
print("ts4 JSON 파일 수:", len(ts4_json_paths))

valid_json_paths = []
invalid_json_paths = []
no_image_json_paths = []

for src_path in tqdm(ts4_json_paths, desc="Checking ts4 JSON & image existence"):
    try:
        with open(src_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except UnicodeDecodeError:
        # 깨진/숨김 파일은 완전히 무시
        invalid_json_paths.append(src_path)
        continue
    except json.JSONDecodeError:
        invalid_json_paths.append(src_path)
        continue
    except Exception as e:
        print(f"[warn] {src_path} 로드 실패: {e}")
        invalid_json_paths.append(src_path)
        continue

    if "images" not in data or len(data["images"]) == 0:
        no_image_json_paths.append(src_path)
        continue

    # 조합 폴더명: .../K-조합_json/K-약코드/파일.json 구조라고 가정
    src_p = Path(src_path)
    combo_dir = src_p.parent.parent.name      # 예: K-000250-000573-002483-006192_json
    combo_name = combo_dir.replace("_json", "")  # 예: K-000250-000573-002483-006192

    exist_count = 0
    for img_info in data["images"]:
        fname = img_info.get("file_name")
        if not fname:
            continue

        rel_path = os.path.join(combo_name, fname)  # train_images/K-조합/파일명
        img_path = os.path.join(ADDED_TRAIN_IMG_DIR, rel_path)

        if os.path.exists(img_path):
            exist_count += 1

    if exist_count > 0:
        valid_json_paths.append(src_path)
    else:
        no_image_json_paths.append(src_path)

print("이미지가 실제로 존재하는 ts4 JSON 수:", len(valid_json_paths))
print("이미지가 하나도 없는 ts4 JSON 수:", len(no_image_json_paths))
print("파싱 실패 JSON 수:", len(invalid_json_paths))
````

````python
if len(valid_json_paths) == 0:
    print("TS4 데이터셋에서 JSON과 실제 이미지가 일치하는 샘플이 없어, 이후 단계(카테고리 수정/56클래스 필터/복사)는 수행하지 않습니다.")
````

### 체크 포인트

- **해당 코드 실행 결과, 이미지가 실제로 존재하는 JSON 수가 0이라면 해당 데이터셋은 사용하지 않는다.**

- TS4 데이터셋의 경우 실행 결과 0 으로 종료하였다.

### 2. TS4 valid JSON을 dl_idx/dl_name 규칙으로 수정하여 별도 폴더에 저장

````python
fixed_cnt = 0
skipped_cnt = 0
error_cnt = 0

for src_path in tqdm(valid_json_paths, desc="Fixing ts4 JSON (dl_idx/dl_name 규칙 적용)"):
    try:
        with open(src_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        print(f"[warn] {src_path} 로드 실패: {e}")
        error_cnt += 1
        continue

    if "images" not in data or len(data["images"]) == 0:
        skipped_cnt += 1
        continue

    img0 = data["images"][0]
    dl_idx = img0.get("dl_idx")
    dl_name = img0.get("dl_name")

    if dl_idx is None or dl_name is None:
        skipped_cnt += 1
        continue

    try:
        cat_id = int(dl_idx)
    except ValueError:
        print(f"[warn] dl_idx 정수 변환 실패: {src_path}, dl_idx={dl_idx}")
        skipped_cnt += 1
        continue

    data["categories"] = [
        {
            "supercategory": "pill",
            "id": cat_id,
            "name": dl_name,
        }
    ]

    if "annotations" in data:
        for anno in data["annotations"]:
            anno["category_id"] = cat_id

    rel_path = os.path.relpath(src_path, ADDED_TRAIN_ANNO_DIR)
    dst_path = os.path.join(ADDED_TRAIN_ANNO_FIXED_DIR, rel_path)
    os.makedirs(os.path.dirname(dst_path), exist_ok=True)

    with open(dst_path, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

    fixed_cnt += 1

print(f"[done] ts4 JSON (valid only) 수정 완료 수: {fixed_cnt}, 건너뜀: {skipped_cnt}, 에러: {error_cnt}")
print("수정된 ts4 JSON 루트 폴더:", ADDED_TRAIN_ANNO_FIXED_DIR)
````

### 3. raw 56개 카테고리 목록 추출

````python
raw_train_annotation_full_paths = find_json_files_recursively(RAW_TRAIN_ANNO_DIR)
print("[raw] 어노테이션 JSON 파일 수:", len(raw_train_annotation_full_paths))

raw_all_category_mappings = []
raw_parsed_category_id_name_pairs_set = set()

for full_path in tqdm(raw_train_annotation_full_paths, desc="[raw] Parsing categories"):
    try:
        with open(full_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        print(f"[raw] 경고: {full_path} 로드 오류: {e}. 건너뜁니다.")
        continue

    if "categories" in data:
        for cat_info in data["categories"]:
            pair = (cat_info["id"], cat_info["name"])
            if pair not in raw_parsed_category_id_name_pairs_set:
                raw_all_category_mappings.append(cat_info)
                raw_parsed_category_id_name_pairs_set.add(pair)

raw_categories_df = pd.DataFrame(raw_all_category_mappings)
raw_global_categories_df = raw_categories_df[["id", "name"]].copy()
raw_global_categories_df.rename(
    columns={"id": "global_category_id", "name": "global_class_name"},
    inplace=True,
)

print("[raw] 고유 클래스 수:", len(raw_global_categories_df))
display(raw_global_categories_df.head())
````

### 4. TS4 fixed JSON (valid only) 파싱 + raw 56개에 해당하는 샘플 필터링

````python
fixed_ts4_json_paths = find_json_files_recursively(ADDED_TRAIN_ANNO_FIXED_DIR)
print("[added-fixed] ts4 fixed JSON 파일 수:", len(fixed_ts4_json_paths))

all_images_meta = []
all_annotations = []
all_category_mappings = []

parsed_image_filenames_set = set()
parsed_annotation_ids_set = set()
parsed_category_id_name_pairs_set = set()

print("[added-fixed] ts4 fixed 어노테이션 파싱 중...")
for full_path in tqdm(fixed_ts4_json_paths, desc="[added-fixed] Parsing Annotations"):
    try:
        with open(full_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        print(f"[added-fixed] 경고: {full_path} 로드 오류: {e}. 건너뜁니다.")
        continue

    if "images" in data:
        for img_info in data["images"]:
            if img_info["file_name"] not in parsed_image_filenames_set:
                img_info["source_json_file"] = full_path
                all_images_meta.append(img_info)
                parsed_image_filenames_set.add(img_info["file_name"])

    if "annotations" in data:
        for anno_info in data["annotations"]:
            if (
                "bbox" in anno_info
                and anno_info["bbox"]
                and len(anno_info["bbox"]) == 4
                and anno_info["id"] not in parsed_annotation_ids_set
            ):
                anno_info["source_json_file"] = full_path
                all_annotations.append(anno_info)
                parsed_annotation_ids_set.add(anno_info["id"])

    if "categories" in data:
        for cat_info in data["categories"]:
            pair = (cat_info["id"], cat_info["name"])
            if pair not in parsed_category_id_name_pairs_set:
                all_category_mappings.append(cat_info)
                parsed_category_id_name_pairs_set.add(pair)

images_df = pd.DataFrame(all_images_meta)
annotations_df = pd.DataFrame(all_annotations)
categories_df = pd.DataFrame(all_category_mappings)

print("\n[added-fixed] 파싱 완료")
print("[added-fixed] images_df 수:", len(images_df))
print("[added-fixed] annotations_df 수:", len(annotations_df))
print("[added-fixed] categories_df 수:", len(categories_df))

# annotations 에 images 메타데이터 병합
annotations_df = pd.merge(
    annotations_df,
    images_df[
        [
            "id",
            "file_name",
            "dl_name",
            "drug_shape",
            "color_class1",
            "back_color",
            "light_color",
            "print_front",
            "dl_idx",
        ]
    ],
    left_on="image_id",
    right_on="id",
    how="left",
    suffixes=("_anno", "_img"),
)
annotations_df.drop(columns=["id_img"], inplace=True)

global_categories_df = categories_df[["id", "name"]].copy()
global_categories_df.rename(
    columns={"id": "global_category_id", "name": "global_class_name"},
    inplace=True,
)
print("[added-fixed] 고유 클래스 수:", len(global_categories_df))
display(global_categories_df.head())
````

### raw 56개에 해당하는 TS4 샘플만 필터링

````python
# raw 56개 클래스 ID 집합
raw_target_ids = set(raw_global_categories_df["global_category_id"].astype(int))
print("[filter] raw 56개 클래스 ID 수:", len(raw_target_ids))

# ts4 fixed images_df 에서 dl_idx 기준으로 필터링
images_df["dl_idx"] = images_df["dl_idx"].astype(int)
added_filtered_images_df = images_df[images_df["dl_idx"].isin(raw_target_ids)].copy()
print("[filter] raw 56개 클래스에 해당하는 ts4-fixed 이미지 수:", len(added_filtered_images_df))

valid_image_ids_for_56 = set(added_filtered_images_df["id"])
added_filtered_annotations_df = annotations_df[
    annotations_df["image_id"].isin(valid_image_ids_for_56)
].copy()
print("[filter] raw 56개 클래스에 해당하는 ts4-fixed 어노테이션 수:", len(added_filtered_annotations_df))
````

### 5. 필터된 JSON/이미지를 fixed 폴더로 복사

- 이미지 존재 여부는 1단계에서 이미 한 번 필터링했기 때문에, 여기서는 대부분 존재해야 한다.

- 그래도 안전하게 한 번 더 체크하면서 복사한다.

````python
# 5-1. 사용된 JSON 파일 집합
used_json_paths = set(added_filtered_images_df["source_json_file"].unique())
print("[copy] 사용된 TS4-fixed JSON 파일 수:", len(used_json_paths))
````

````python
# 5-2. JSON 복사

json_copied_cnt = 0

for src_path in tqdm(used_json_paths, desc="[copy] TS4-fixed JSON -> fixed/train_annotations"):
    rel_path = os.path.relpath(src_path, ADDED_TRAIN_ANNO_FIXED_DIR)
    dst_path = os.path.join(FIXED_TRAIN_ANNO_DIR, rel_path)
    os.makedirs(os.path.dirname(dst_path), exist_ok=True)

    shutil.copy2(src_path, dst_path)
    json_copied_cnt += 1

print("[copy] 복사된 어노테이션 JSON 개수:", json_copied_cnt)
````

````python
# 5-3. 이미지 복사

img_copied_cnt = 0
img_missing_cnt = 0
img_skipped_cnt = 0

for _, row in added_filtered_images_df.iterrows():
    fname = row["file_name"]
    src_json = Path(row["source_json_file"]).resolve()

    combo_dir = src_json.parent.parent.name
    combo_name = combo_dir.replace("_json", "")

    # 실제 TS4 이미지 상대 경로: K-조합/파일명
    rel_path = os.path.join(combo_name, fname)

    src = os.path.join(ADDED_TRAIN_IMG_DIR, rel_path)
    dst = os.path.join(FIXED_TRAIN_IMG_DIR, rel_path)

    if not os.path.exists(src):
        print(f"[copy warning] 원본 이미지 없음: {src}")
        img_missing_cnt += 1
        continue

    os.makedirs(os.path.dirname(dst), exist_ok=True)

    if os.path.exists(dst):
        img_skipped_cnt += 1
        continue

    shutil.copy2(src, dst)
    img_copied_cnt += 1

print(f"[copy] 새로 복사된 이미지: {img_copied_cnt}개, 이미 존재해서 건너뜀: {img_skipped_cnt}개, 원본 없음: {img_missing_cnt}개")
````

````python
# 추가된 데이터(TS4 기반 fixed)의 카테고리 목록, 카테고리별 JSON 수, 이미지 수 확인

from collections import defaultdict

# 1) fixed 어노테이션 JSON 목록 수집
fixed_json_paths = find_json_files_recursively(FIXED_TRAIN_ANNO_DIR)
print("[fixed] 어노테이션 JSON 파일 수:", len(fixed_json_paths))

# 2) 카테고리별 집계용 딕셔너리
cat_name_map = {}                      # category_id -> category_name
cat_json_set = defaultdict(set)        # category_id -> {json_path}
cat_image_set = defaultdict(set)       # category_id -> {file_name}

for json_path in tqdm(fixed_json_paths, desc="[fixed] 카테고리/JSON/이미지 집계"):
    try:
        with open(json_path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        print(f"[fixed warn] {json_path} 로드 실패: {e}")
        continue

    # categories 섹션에서 id/name 추출
    cat_id_to_name = {}
    for cat in data.get("categories", []):
        cid = cat.get("id")
        cname = cat.get("name")
        if cid is not None:
            cat_id_to_name[int(cid)] = cname
            # 전체 맵에도 기록
            if cid not in cat_name_map:
                cat_name_map[cid] = cname

    # image_id -> file_name 매핑
    image_id_to_fname = {}
    for img in data.get("images", []):
        iid = img.get("id")
        fname = img.get("file_name")
        if iid is not None and fname:
            image_id_to_fname[int(iid)] = fname

    # annotations 에서 category_id / image_id 사용
    for anno in data.get("annotations", []):
        cid = anno.get("category_id")
        iid = anno.get("image_id")
        if cid is None or iid is None:
            continue
        cid = int(cid)
        fname = image_id_to_fname.get(int(iid))
        # JSON 경로 집계
        cat_json_set[cid].add(json_path)
        # 이미지 파일명 집계
        if fname:
            cat_image_set[cid].add(fname)

# 3) 결과 DataFrame으로 정리
rows = []
for cid, cname in cat_name_map.items():
    json_count = len(cat_json_set.get(cid, set()))
    img_count = len(cat_image_set.get(cid, set()))
    rows.append(
        {
            "category_id": cid,
            "category_name": cname,
            "num_json_files": json_count,
            "num_images": img_count,
        }
    )

fixed_cat_summary_df = pd.DataFrame(rows).sort_values("category_id").reset_index(drop=True)

print("[fixed] 추가된 데이터 카테고리 요약:")
display(fixed_cat_summary_df)
````

````python
# fixed 데이터셋에서 이미지 + 어노테이션 샘플 시각화
def load_fixed_coco(fixed_json_path):
    """단일 COCO 형식 JSON (fixed) 를 불러오는 헬퍼 함수 (필요 시 사용)."""
    with open(fixed_json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data

def visualize_fixed_sample(
    fixed_json_paths=None,
    num_samples=3,
    figsize=(6, 6),
):
    """
    fixed/train_annotations 폴더에서 임의 JSON을 골라,
    해당 JSON 안의 이미지+어노테이션을 샘플 시각화.
    """
    if fixed_json_paths is None:
        fixed_json_paths = find_json_files_recursively(FIXED_TRAIN_ANNO_DIR)

    if len(fixed_json_paths) == 0:
        print("[viz] fixed/train_annotations 에 JSON 이 없습니다.")
        return

    # 시각화할 JSON 몇 개 랜덤 선택
    sample_jsons = random.sample(
        fixed_json_paths, min(num_samples, len(fixed_json_paths))
    )

    for json_path in sample_jsons:
        print("\n[viz] JSON:", json_path)

        try:
            with open(json_path, "r", encoding="utf-8") as f:
                data = json.load(f)
        except Exception as e:
            print(f"[viz warn] {json_path} 로드 실패: {e}")
            continue

        # image_id -> image info (file_name, width, height)
        img_dict = {img["id"]: img for img in data.get("images", [])}
        # category_id -> name
        cat_dict = {cat["id"]: cat.get("name", str(cat["id"])) for cat in data.get("categories", [])}

        # 이미지별 어노테이션 묶기
        img_to_annos = {}
        for anno in data.get("annotations", []):
            iid = anno.get("image_id")
            if iid not in img_to_annos:
                img_to_annos[iid] = []
            img_to_annos[iid].append(anno)

        # 이 JSON 안에서 시각화할 이미지 몇 개 선택
        img_ids = list(img_dict.keys())
        if not img_ids:
            print("[viz] 해당 JSON 에 images 가 없습니다.")
            continue

        sample_img_ids = random.sample(img_ids, min(3, len(img_ids)))

        for iid in sample_img_ids:
            img_info = img_dict[iid]
            fname = img_info["file_name"]
            width = img_info.get("width")
            height = img_info.get("height")

            # source_json_file 로부터 combo_name 계산
            # 이 JSON 은 이미 fixed 위치이므로, 원본 TS4 구조와 동일한 상위 폴더 이름을 사용
            src_json = Path(json_path)
            combo_dir = src_json.parent.parent.name  # 예: K-000250-000573-002483-006192_json
            combo_name = combo_dir.replace("_json", "")

            # 실제 이미지 경로: fixed 가 아니라 원본 TS4 이미지에서 바로 보는 경우
            # src_img_path = os.path.join(ADDED_TRAIN_IMG_DIR, combo_name, fname)
            # fixed/train_images 에서 보는 경우:
            src_img_path = os.path.join(FIXED_TRAIN_IMG_DIR, combo_name, fname)

            if not os.path.exists(src_img_path):
                print(f"[viz warn] 이미지 파일 없음: {src_img_path}")
                continue

            # 이미지 로드
            img = Image.open(src_img_path).convert("RGB")

            # 시각화
            fig, ax = plt.subplots(1, 1, figsize=figsize)
            ax.imshow(img)
            ax.set_title(f"{fname}\n(json: {Path(json_path).name})")
            ax.axis("off")

            # 어노테이션 박스 그리기
            annos = img_to_annos.get(iid, [])
            for anno in annos:
                bbox = anno.get("bbox")
                cid = anno.get("category_id")
                cname = cat_dict.get(cid, str(cid))

                if not bbox or len(bbox) != 4:
                    continue

                x, y, w, h = bbox
                rect = patches.Rectangle(
                    (x, y),
                    w,
                    h,
                    linewidth=2,
                    edgecolor="red",
                    facecolor="none",
                )
                ax.add_patch(rect)
                ax.text(
                    x,
                    y - 2,
                    f"{cname} ({cid})",
                    fontsize=8,
                    color="yellow",
                    bbox=dict(facecolor="red", alpha=0.5, edgecolor="none"),
                )

            plt.show()
````
