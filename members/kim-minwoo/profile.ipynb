## Data Analysis Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 샘플 데이터 생성
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=100)
data = {
    'date': dates,
    'sales': np.random.randint(100, 1000, 100),
    'customers': np.random.randint(10, 100, 100)
}
df = pd.DataFrame(data)

# 기본 통계
print("데이터 요약:")
print(df.describe())
print(f"\n총 매출: {df['sales'].sum():,}원")
print(f"평균 고객 수: {df['customers'].mean():.1f}명")
```

```python
# 시각화
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# 매출 추이
axes[0].plot(df['date'], df['sales'], color='blue', linewidth=2)
axes[0].set_title('일별 매출 추이', fontsize=14, fontweight='bold')
axes[0].set_xlabel('날짜')
axes[0].set_ylabel('매출 (원)')
axes[0].grid(True, alpha=0.3)

# 고객 수 추이
axes[1].plot(df['date'], df['customers'], color='green', linewidth=2)
axes[1].set_title('일별 고객 수 추이', fontsize=14, fontweight='bold')
axes[1].set_xlabel('날짜')
axes[1].set_ylabel('고객 수 (명)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
